# **Data Cleaning**

```python
# Count missing values
missing_values = data.isnull().sum()

# Calculate percentage of missing values
missing_percentage = (missing_values / len(data)) * 100

# Combine into a DataFrame for better readability
missing_summary = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
print(missing_summary)
```

```python
# droping the columns which has 100% null values
data.dropna(axis=1, how='all', inplace=True)
```

```python
data.isna().sum()/len(data) * 100
```

```python
# dropping the columns which has more than 60% null values
#also dropping desc as it is a Loan description provided by the borrower which has 36% missing values
# but we also have title column which is The loan title provided by the borrower so instead of desc we can use title
# Check if the columns exist before dropping
columns_to_drop = ['next_pymnt_d', 'mths_since_last_record', 'mths_since_last_delinq', 'desc']
existing_columns = [col for col in columns_to_drop if col in data.columns]

data.drop(columns=existing_columns, inplace=True)
```

```python
data.isna().sum()/len(data) * 100
```

```python
new = pd.read_csv('/content/cleaned_loan_data.csv')
```

```python
# Visualizing loan term distribution
sns.countplot(x='term', data=new)
plt.title('Distribution of Loan Term')
plt.xlabel('Loan Term')
plt.ylabel('Count')
plt.show()
```

**Observation** :- The countplot shows the distribution of loan terms. It appears that loans with a term of 36 months are significantly more common than loans with a term of 60 months.

```python
sns.boxplot(x = 'term', y = 'int_rate', data=new)
```

**Observation** :-The median interest rate for 60-month loans is noticeably higher than the median interest rate for 36-month loans.

```python
# Visualizing loan amount distribution
sns.histplot(new['loan_amnt'], kde=True)
plt.title('Distribution of Loan Amount')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

**Observation**

*The histogram is right-skewed, indicating that most loans are concentrated towards the lower end of the loan amount range.

*There are notable peaks around specific loan amounts, such as $10,000, $$10,000, $15,000, and $20,000. This suggests that these are common loan amounts requested by borrowers.

```python
loan_status_counts = new['loan_status'].value_counts()

# Create the pie chart
plt.pie(loan_status_counts, labels=loan_status_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Loan Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

**Observations**:

*A large majority of loans have been fully paid.

*A significant portion of loans have been charged off, indicating defaults.

*A small percentage of loans are currently ongoing.

```python
# Create the histogram and KDE plot
sns.histplot(new['revol_bal'], kde=True)

# Set plot title and labels
plt.title('Distribution of Credit Balance')
plt.xlabel('Credit Balance (revol_bal)')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

The distribution of credit balances (revol_bal) is heavily right-skewed. This indicates that most borrowers have relatively low credit balances, while a smaller number of borrowers have very high credit balances.

```python
# Create the box plot
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
sns.boxplot(x='purpose', y='loan_amnt', data=new , hue = 'purpose', palette='coolwarm')
plt.title('Loan Amount by Purpose')
plt.xlabel('Purpose')
plt.ylabel('Loan Amount')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
purpose_counts = new["purpose"].value_counts()
plt.figure(figsize=(10, 4))
sns.barplot(x=purpose_counts.values, y=purpose_counts.index, palette="viridis")

# Add titles and labels
plt.title('Frequency of Unique Values in Purpose Column')
plt.xlabel('Count')
plt.ylabel('Purpose')
plt.xticks(rotation=0)


# Show the plot
plt.show()
```

The most frequent purpose for taking a loan is debt_consolidation, followed by credit_card. This suggests that a significant portion of loans are taken out to consolidate existing debt or to pay off credit card balances.

```python
# Summary statistics for 'int_rate'
print(new['int_rate'].describe())

# Plot the distribution
plt.figure(figsize=(8, 5))
sns.histplot(new['int_rate'], bins=30, kde=True, color='orange')
plt.title('Distribution of Interest Rates')
plt.xlabel('Interest Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Summary statistics for 'annual_inc'
print(new['annual_inc'].describe())

# Plot the distribution of annual income
plt.figure(figsize=(8, 5))
sns.histplot(new['annual_inc'], bins=50, kde=True, color='green')
plt.title('Distribution of Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.show()

# Identify income outliers using the 1.5x IQR rule
Q1 = new['annual_inc'].quantile(0.25)
Q3 = new['annual_inc'].quantile(0.75)
IQR = Q3 - Q1
outliers = new[(new['annual_inc'] < (Q1 - 1.5 * IQR)) | (new['annual_inc'] > (Q3 + 1.5 * IQR))]

print(f"Number of outliers in 'annual_inc': {len(outliers)}")
```

The histogram would likely show that the distribution of annual income is right-skewed. This is common with income data, where most people have lower to middle incomes, and a smaller number of people have very high incomes.

```python
# Group by 'emp_length' and calculate average loan amount
avg_loan_by_emp_length = new.groupby('emp_length')['loan_amnt'].mean().sort_values(ascending=False)

print("Average Loan Amount by Employment Length:")
print(avg_loan_by_emp_length)

# Plot the average loan amount by employment tenure
plt.figure(figsize=(8, 5))
avg_loan_by_emp_length.plot(kind='bar', color='blue')
plt.title('Average Loan Amount by Employment Length')
plt.xlabel('Employment Length')
plt.ylabel('Average Loan Amount')
plt.show()
```

This suggests that lenders might consider employment length as a factor in determining loan amounts, potentially because longer tenure indicates greater financial stability. Individuals with 10+ years of employment history have the highest average loan amounts.

```python
# Group by 'home_ownership' and calculate average income
avg_income_by_home_ownership = new.groupby('home_ownership')['annual_inc'].mean().sort_values(ascending=False)

print("Average Annual Income by Home Ownership Type:")
print(avg_income_by_home_ownership)

# Plot the average income by home ownership type
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_income_by_home_ownership.index, y=avg_income_by_home_ownership.values, palette='muted')
plt.title('Average Annual Income by Home Ownership Type')
plt.xlabel('Home Ownership')
plt.ylabel('Average Annual Income')
plt.show()
```

The graph show that borrowers who own their homes (represented by 'OWN' or 'MORTGAGE') have, on average, higher annual incomes compared to those who rent ('RENT') or have other housing arrangements ('OTHER', 'NONE').

```python
# Group by 'zip_code' and calculate average loan amount
avg_loan_by_zip_code = new.groupby('zip_code')['loan_amnt'].mean().sort_values(ascending=False)

print("Average Loan Amount by Zip Code (Top 10):")
print(avg_loan_by_zip_code.head(10))

# Plot the top 10 zip codes with the highest average loan amount
top_zip_codes = avg_loan_by_zip_code.head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_zip_codes.index, y=top_zip_codes.values, palette='coolwarm')
plt.title('Top 10 Zip Codes by Average Loan Amount')
plt.xlabel('Zip Code')
plt.ylabel('Average Loan Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Remove the '%' symbol and convert the column to float
new['int_rate'] = new['int_rate'].str.replace('%', '').astype(float)

# Verify the data type
print(new['int_rate'].dtype)

# Calculate the average interest rate for each grade
avg_int_rate_by_grade = new.groupby('grade')['int_rate'].mean().sort_values(ascending=False)

print("Average Interest Rate by Grade:")
print(avg_int_rate_by_grade)

# Plot the average interest rate by grade
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_int_rate_by_grade.index, y=avg_int_rate_by_grade.values, palette='coolwarm')
plt.title('Average Interest Rate by Lending Grade')
plt.xlabel('Grade')
plt.ylabel('Average Interest Rate (%)')
plt.show()
```

```python
verification_status_counts = new['verification_status'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(verification_status_counts, labels=verification_status_counts.index, autopct='%1.1f')

plt.title('Verification Status Distribution')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='verification_status', y='loan_amnt', data=new, palette='viridis')
plt.title("Loan Amount Distribution by Verification Status")
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.kdeplot(data=new, x='int_rate', hue='verification_status', fill=True, common_norm=False, palette='Set2')
plt.title("Interest Rate Distribution by Verification Status (KDE)")
plt.xlabel("Interest Rate (%)")
plt.ylabel("Density")
plt.show()
```

```python
# Scatter plot to visualize the relationship
plt.figure(figsize=(8, 5))
sns.scatterplot(x='loan_amnt', y='installment', data=new, alpha=0.6, color='purple')
plt.title('Relationship Between Loan Amount and Installment')
plt.xlabel('Loan Amount')
plt.ylabel('Installment')
plt.show()

# Calculate the correlation between loan amount and installment
correlation = new['loan_amnt'].corr(new['installment'])
print(f"Correlation between Loan Amount and Installment: {correlation:.2f}")
```

```python
# Define default statuses (Default or Charged Off)
default_statuses = ['Default', 'Charged Off']
new['is_default'] = new['loan_status'].isin(default_statuses).astype(int)

# Plot the distribution of int_rate for default vs non-default loans
plt.figure(figsize=(8, 5))
sns.boxplot(x='is_default', y='int_rate', data=new, hue='is_default', palette='coolwarm')
plt.title('Interest Rate vs Default Status')
plt.xlabel('Default Status (0: No, 1: Yes)')
plt.ylabel('Interest Rate (%)')
plt.show()

# Calculate the average int_rate for defaulted and non-defaulted loans
avg_int_rate_default = new.groupby('is_default')['int_rate'].mean()
print("Average Interest Rate by Default Status:")
print(avg_int_rate_default)
```

```python
# Calculate default rates for each grade
default_rates_by_grade = new.groupby('grade')['is_default'].mean().sort_values(ascending=False)

print("Default Rates by Grade:")
print(default_rates_by_grade)

# Plot default rates by grade
plt.figure(figsize=(8, 5))
sns.barplot(x=default_rates_by_grade.index, y=default_rates_by_grade.values, palette='viridis')
plt.title('Default Rates by Lending Grade')
plt.xlabel('Grade')
plt.ylabel('Default Rate')
plt.show()
```

```python
# Plot the distribution of revol_bal
plt.figure(figsize=(8, 5))
sns.histplot(new['revol_bal'], bins=50, kde=True, color='blue')
plt.title('Distribution of Revolving Balances')
plt.xlabel('Revolving Balance')
plt.ylabel('Frequency')
plt.show()

# Calculate the total revolving balance by state
revol_bal_by_state = new.groupby('addr_state')['revol_bal'].sum().sort_values(ascending=False)

print("Total Revolving Balances by State:")
print(revol_bal_by_state)

# Plot the total revolving balances by state
plt.figure(figsize=(15, 6))
sns.barplot(x=revol_bal_by_state.index, y=revol_bal_by_state.values, palette='Purples')
plt.title('Total Revolving Balances by State')
plt.xlabel('State')
plt.ylabel('Total Revolving Balances')
plt.xticks(rotation=90)
plt.show()
```

```python
# Identifying key correlations
corr = new.loc[:, [ 'funded_amnt', 'funded_amnt_inv', 'loan_amnt']].corr()
plt.figure(figsize = (8,6))
sns.heatmap(corr, annot = True, cmap='Blues')
plt.show()
```
